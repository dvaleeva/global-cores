```r
# Snapshot world 2009 February
# Not dynamic data
# Deleted individuals with > 100 positions

# Weighted k-shell
# Using https://github.com/antongrau/eliter/blob/master/R/core%20identification.R
```

```r
library(eliter)
library(tidyverse)
library(plyr)
```

```r
df1 <- read.csv('~/Dropbox/global cores/world/2009feb_core.csv', header = TRUE, sep = ",")
nrow(df1)
```

```r
head(df1)
```

```r
graph.all <- elite.network(df1)
summary(graph.all)
```

```r
graph.com <- largest.component(graph.all)
summary(graph.com)
```

```r
sp <- distances(graph.com)
# calculates the length of all the shortest paths from or to the vertices in the network
```

```r
head(sp)
```

```r
sp[sp >= 2.1] <- 0 # if shortest path more than 2, assign 0 

graph.reach <- graph_from_adjacency_matrix(Matrix(sp), mode = "undirected", weighted = TRUE)

summary(graph.reach)
```

```r
shell <- k.shell(graph.reach, start.level = 0, verbose = TRUE)
```
